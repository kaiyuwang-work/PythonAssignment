## Count Layers

```python
items = ''
def countLayers(search_content):
    global items
    layers = 0 
    items = gis.content.search(search_content,'feature layer')
    for item in items:
        print(item)
        for lyr in item.layers:
            print(lyr.properties.name)
            layers+=1
    return layers
    
print(countLayers('owner:macdo925_utoronto')," layers") 
```

## List Layers

```python
items = ''
def listLayers(search_content):
    global items
    items = gis.content.search(search_content,'feature layer')
    for item in items:
        print('item ',item.title)
        for lyr in item.layers:
            print('\t'+lyr.properties.name)
        print("\n")
    
print(listLayers('owner:macdo925_utoronto')) 
```

## Initialize a Map currently focusing on Toronto

```python
def initializeMap(address):
    mapx = gis.map(address)
    return mapx
mapx = initializeMap('toronto,ontario')
```

## Add Item to a Map currently

```python
def addItemToMap(items, mapx):
    for item in items:
        mapx.add_layer(item)
```

## Add A Layer to a Map currently

```python
def addLayerToMap(layer, mapx):
    mapx.add_layer(layer)
```

```python
utoronto = items[2].layers[0]
addLayerToMap(utoronto,mapx)
mapx
```

## Remove Layer

```python
def removeLayer(layer,mapx):
    mapx.remove_layers(layers=layer)
removeLayer(utoronto,mapx)
```

## Draw Neighborhood Population 

```python
def drawBarPlot(df):
    df = df.sort_values(by="AvgTotPop" , ascending=False) 
    %matplotlib inline

    n = len(df)/2
    x = np.arange(0, n, step=.5)
    y = x ** 2

    norm = plt.Normalize(y.min(), y.max())
    norm_y = norm(y)
    
    plt.rcParams['figure.figsize'] = (30,50)
    map_vir = cm.get_cmap(name='plasma')
    color = map_vir(norm_y)
    plt.title("Neighborhood Population Barplot",fontsize=50)
    #plt.xticks(rotation=270)
    
    plt.xticks(fontsize=30)
    plt.yticks(fontsize=16)
    plt.barh(range(len(df['AvgTotPop'])), df['AvgTotPop'],tick_label=df['NAME'],color=color)
    #plt.set_size_inches(18.5,10.5)
    plt.show()
    
    
```

```python
df = items[0].layers[0].query().sdf
drawBarPlot(df)
```

## Calculate Layer Attribute

```python
def attributesInLayer(layer):
    return layer.query().sdf.shape[1]
```

```python
print(attributesInLayer(items[0].layers[0]), "attributes in this layer.")
```

## Calculate Layer Rows

```python
def rowsInLayer(layer):
    return layer.query().sdf.shape[0]
```

```python
print(rowsInLayer(items[0].layers[0]), "rows in this layer.")
```

## Summarize Layer Size

```python
def summarizeLayer(layer):
    print("This layer has "+ str(attributesInLayer(layer))+" and "+str(rowsInLayer(layer))+" rows.")
summarizeLayer(items[0].layers[0])
```
