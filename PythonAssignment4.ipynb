```python
import time
import re

#welcome message
def displayWelcomeMessage():
        print('Hi there. this is Kaiyu Wang, today is '+time.strftime("%Y-%m-%d %H:%M:%S", time.localtime())+', and I will guide you to calculate windchill later :-)')
displayWelcomeMessage()


#calculate windchill
def CalculateWindChill(temperature, windspeed):
    windchill = 13.12+(0.6215*temperature)-(11.37*windspeed**0.16)+(0.3965*temperature*windspeed**0.16)
    return windchill


#make sure user is inputing valid numbers rather than string
def popup(question):
    result = input(question)
    while re.match(r'-*[0-9]*', result).group(0) != result:
        result = input("Please try to input a number")
    return result



def popup2(question):
    start = input("Enter the starting "+question) 
    #make sure it's a number and that can be converted to int
    while re.match(r'-*[0-9]*', start).group(0) != start:
        start = input("Please try to input a number")
    end = input("Enter the ending "+question)
    # make sure it's a number and should be bigger than start point
    while re.match(r'-*[0-9]*', end).group(0) != end or int(end)<=int(start):
        end = input("Please try to input a number and no less than start number ")
    step = input("Enter the step increment for the "+question)
    while re.match(r'-*[0-9]*', step).group(0) != step:
        step = input("Please try to input a number and no bigger than range gap")
    return (int(start), int(end), int(step))

#ask for variable values and outprint windchill result    
def Outprint():
    temp = popup2("temperature")
    sTemp = temp[0]
    eTemp = temp[1]
    stepTemp = temp[2]
    temp = popup2("wind speed")
    sWind = temp[0]
    eWind = temp[1]
    stepWind = temp[2]
    
    print("\n\n---------Here's your windchill calculation-------------\n")
    
    for temperature in range(sTemp,eTemp,stepTemp):
        print("temperature is %d" % temperature)
        for windspeed in range(sWind,eWind,stepWind):
            print("\twindspeed is "  + str(windspeed)+" caculated wind chill is: "+ str(int(CalculateWindChill(temperature, windspeed))))
Outprint()
```
